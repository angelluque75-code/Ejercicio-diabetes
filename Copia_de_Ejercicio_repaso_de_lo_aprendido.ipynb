# Primera parte del ejercicio. 1. Crear un dataset sencillo con NumPy

Vamos: a generar datos numéricos continuos, discretos y categóricos para entender cómo se representan en Python.

```python
import numpy as np

# Datos numéricos continuos (temperatura)
temperaturas = np.random.uniform(18, 30, size=30)

# Datos numéricos discretos (edad)
edades = np.random.randint(18, 65, size=30)

# Datos categóricos nominales (colores favoritos)
colores = np.random.choice(['Rojo', 'Azul', 'Verde', 'Amarillo'], size=30)

temperaturas[:5], edades[:5], colores[:5]
```

Como has visto en la celda anterior, en este ejercicio comenzamos importando la librería NumPy, porque es la herramienta estándar en Python para trabajar con datos numéricos de manera eficiente. NumPy nos permite crear arrays, generar números aleatorios y realizar operaciones matemáticas muy rápidas, algo fundamental tanto en analítica como en ciencia de datos.

Cuando escribimos import numpy as np, lo que hacemos es cargar la librería y abreviarla como np para poder usar sus funciones sin escribir nombres largos. A partir de aquí, todo lo que empiece por np. viene de NumPy.

Después utilizamos np.random.uniform, que es una función especializada de NumPy para generar números aleatorios continuos dentro de un rango. La usamos porque queremos simular temperaturas reales, que no son enteras, sino valores medibles con decimales.
Esto nos permite crear datos que se parecen a un conjunto real de temperaturas sin tener que tomarlas manualmente.

A continuación usamos np.random.randint, otra función de la misma familia, pero especializada en generar números enteros (discretos). Aquí queremos simular edades, y la edad siempre es un número entero.
Es la forma más simple y rápida de simular un conjunto de datos demográficos básicos.

Finalmente, utilizamos np.random.choice, que es una función de NumPy diseñada para seleccionar categorías de forma aleatoria. La aplicamos para crear una columna con colores favoritos, que es un ejemplo típico de datos categóricos nominales.

Al final, mostramos una pequeña parte de los datos generados. Esto permite ver los primeros 5 valores de cada uno, lo que sirve para comprobar que la generación de datos se ha realizado correctamente.

# 2. Crear un DataFrame con Pandas

Ahora vamos a ver cómo integrar tipos de datos distintos en una estructura tabular.

```python
import pandas as pd

df = pd.DataFrame({
    "Temperatura": temperaturas,
    "Edad": edades,
    "ColorFavorito": colores
})

df.head()
```

Aquí estamos importando la librería Pandas, que es la herramienta fundamental para manejar datos en formato tabular en Python. Pandas nos permite crear tablas, leer archivos CSV, limpiar datos, agrupar, filtrar y trabajar cómodamente con datasets como si fueran hojas de cálculo o tablas SQL.
El alias pd es simplemente una abreviatura estándar para no escribir pandas entero cada vez.

Después usamos pd.DataFrame, que es la función (o constructor) que utiliza pandas para crear un DataFrame, que es básicamente una tabla con filas y columnas.

Lo que hace esta línea es:

Tomar un diccionario de Python donde cada clave es el nombre de una columna.
Ejemplo: "Temperatura" será el nombre de la columna.

Asignar a cada columna una lista o array de datos que ya hemos generado antes con NumPy:

"Temperatura" → array temperaturas

"Edad" → array edades

"ColorFavorito" → array colores

Pandas toma esas tres columnas, las alinea por índice (0 al 29), y construye una tabla completa.
Lo importante es que todas las columnas tienen la misma longitud, así que pandas puede construir filas válidas.

inalmente llamamos a df.head(), que es una función rápida de Pandas para visualizar las primeras 5 filas del DataFrame.

Esto sirve como una comprobación visual:

Ver si las columnas se han creado correctamente.

Ver si los datos están bien alineados.

Confirmar que la estructura final tiene sentido.

En Colab o Jupyter, esta última línea se muestra automáticamente como una tabla con las primeras filas del dataset.

# 3. Estadísticos básicos con Pandas

Conexión directa al glosario: media, mediana, moda, rango, IQR…

```python
df.describe()
```

Para datos categóricos

```python
df['ColorFavorito'].value_counts()
```

Cuando llegamos a esta parte del ejercicio, ya tenemos un DataFrame formado con nuestras tres columnas: temperatura, edad y color favorito.
Ahora toca analizar esos datos, y para eso pandas trae funciones estadísticas listas para usar.

Aquí estamos usando describe(), una función de pandas que resume automáticamente las estadísticas fundamentales de las columnas numéricas del DataFrame.

¿Para qué sirve?
Para hacer en un segundo lo que un analista haría a mano con una tabla Excel: revisar valores mínimos, máximos, la media e incluso la dispersión.

¿Qué calcula exactamente?

count → cuántos datos hay

mean → la media

std → la desviación típica

min → el valor mínimo

25% → primer cuartil

50% → mediana

75% → tercer cuartil

max → valor máximo

Esto conecta directamente con el glosario: media, mediana, rango, cuartiles e IQR (aunque el IQR no aparece directo, se calcula restando Q3 – Q1). Es decir, Estamos pidiendo a pandas un resumen estadístico automático de nuestros datos numéricos.

# Conteo de categorías
Aquí usamos value_counts(), una función diseñada para trabajar con datos categóricos, especialmente nominales.
Pandas cuenta cuántas veces aparece cada categoría (Rojo, Azul, Verde, Amarillo) dentro de la columna.

¿Para qué sirve esto?

Ver la categoría más frecuente → moda

Detectar desequilibrios

Saber si una muestra está sesgada

Preparar datos para gráficos de barras

En resumen: Estamos calculando la moda y la distribución de frecuencias de un dato categórico.

# 4. Visualizar datos con Matplotlib

Mini-introducción real a gráficos: líneas, histogramas y barras.

##Histograma de temperatura

```python
import matplotlib.pyplot as plt

plt.hist(df['Temperatura'], bins=10)
plt.title("Distribución de Temperaturas")
plt.xlabel("Temperatura")
plt.ylabel("Frecuencia")
plt.show()
```

## Gráfico de barras del color favorito

```python
df['ColorFavorito'].value_counts().plot(kind='bar')
plt.title("Colores Favoritos")
plt.xlabel("Color")
plt.ylabel("Conteo")
plt.show()
```

##Gráfico de dispersión: Edad vs Temperatura

```python
plt.scatter(df['Edad'], df['Temperatura'])
plt.title("Edad vs Temperatura")
plt.xlabel("Edad")
plt.ylabel("Temperatura")
plt.show()
```
